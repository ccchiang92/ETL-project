```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
raw_korea_df=pd.read_csv('../KOSPI Historical Data.csv')
raw_china_df=pd.read_csv('../Shanghai Composite Historical Data.csv')
```

```python
korea_df=raw_korea_df[["Date","Price","Change %"]].copy()
china_df=raw_china_df[["Date","Price","Change %"]].copy()
```

```python
korea_df['Date'] =  pd.to_datetime(korea_df['Date'])
china_df['Date'] =  pd.to_datetime(china_df['Date'])
```
